```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Straight line
theta0 = 30
theta1 = 0.5

x = np.linspace(0, 100, 500)
y = theta0 + theta1 * x

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label=f'y = {theta0} + {theta1}x')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Linear Function: y = $\theta_0 + \theta_1x$')
plt.legend()
plt.grid(True)
plt.text(0.95, 0.05, f'$\\theta_0$ = {theta0}\n$\\theta_1$ = {theta1}', 
         transform=plt.gca().transAxes, verticalalignment='bottom', 
         horizontalalignment='right', bbox=dict(facecolor='white', alpha=0.8))
plt.show()
```

```python
# Quadratic function
theta0 = 20
theta1 = 25

x = np.linspace(0, 50, 500)

y = (x - theta1)**2 + theta0

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r-', label=f'y = (x - {theta1})² + {theta0}')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Quadratic Function: y = $(x - \theta_1)^2 + \theta_0$')
plt.legend()
plt.grid(True)

plt.text(0.95, 0.95, f'$\\theta_0$ = {theta0}\n$\\theta_1$ = {theta1}', 
         transform=plt.gca().transAxes, verticalalignment='top', 
         horizontalalignment='right', bbox=dict(facecolor='white', alpha=0.8))

plt.ylim(0, 800)

plt.show()
```

```python
# Log function
x = np.linspace(0.01, 0.99, 1000)  # Avoid 0 and 1 due to log domain

y1 = -np.log(x)
y2 = -np.log(1-x)

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x, y1, 'b-', label='y = -log(x)')
plt.plot(x, y2, 'r-', label='y = -log(1-x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Log Functions: y = -log(x) and y = -log(1-x)')
plt.legend()
plt.grid(True)

# Set reasonable axis limits
plt.xlim(0, 1)
plt.ylim(0, 5)

# Add vertical asymptotes
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=1, color='gray', linestyle='--', alpha=0.5)

# Add text annotations
plt.text(0.02, 4.5, 'x = 0 asymptote', rotation=90, verticalalignment='top')
plt.text(0.98, 4.5, 'x = 1 asymptote', rotation=90, verticalalignment='top')

plt.show()
```

```python
# Sigmoid function implementation
x = np.linspace(-10, 10, 1000)
y = 1 / (1 + np.exp(-x))

# Create the plot for the sigmoid function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = \frac{1}{1 + e^{-x}}$', color='green')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid Function: $y = \\frac{1}{1 + e^{-x}}$')
plt.grid(True)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, label=r'$y = 0.5$ asymptote')
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5, label=r'$x = 0$ axis')

plt.legend()
plt.show()
```

# Task: Music Popularity Prediction
Predict the potential popularity (measured in expected streaming numbers for the first month) of a song before its release based on various musical and contextual features.

## Inputs

### 1. Audio features:
* Tempo (BPM)
* Key/scale
* Energy level (derived from amplitude and frequency analysis)
* Danceability (rhythm patterns)
* Instrumental complexity
* Vocal characteristics (pitch range, vocal density)
* Genre indicators

## Output
* Predicted number of streams for the first month after release (continuous value)
* This makes it a regression problem since we're predicting a continuous numerical value.

## Data Preparation

### 1. Training Data Collection (70% of total dataset):
* Partner with music streaming platforms to get historical data of songs released in the past 3 years
* Collect audio files of released songs
* Gather metadata from music industry databases
* Extract social media metrics from platforms' APIs
* Compile marketing spend data from record labels
* Size: ~100,000 songs with complete feature sets

### 2. Validation Data (15% of total dataset):
* Select a random subset of more recent songs (last 6 months)
* Ensure representation across different genres and artist popularity levels
* Size: ~20,000 songs

### 3. Testing Data (15% of total dataset):
* Most recent songs (last 3 months)
* Include a mix of major label and independent releases
* Size: ~20,000 songs

## Ground-truth Label Collection
1. Streaming platforms provide actual streaming numbers for the first month after release
2. Ensure data normalization across different platforms (Spotify, Apple Music, etc.)
3. Account for regional variations in streaming numbers
4. Document any unusual events that might have affected streaming numbers (viral moments, controversies, etc.)
